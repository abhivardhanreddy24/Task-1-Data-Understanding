```python
import pandas as pd
import numpy as np

# Load Titanic dataset (online)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("\nDataset loaded successfully!")


# Display First & Last Few Records


print("\nFirst 5 records of the dataset:")
print(df.head())

print("\nLast 5 records of the dataset:")
print(df.tail())


#  Identify Data Types


numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
categorical_features = df.select_dtypes(include=['object']).columns
binary_features = [col for col in df.columns if df[col].nunique() == 2]

print("\nNumerical Features:")
print(list(numerical_features))

print("\nCategorical Features:")
print(list(categorical_features))

print("\nBinary Features:")
print(binary_features)


#  Dataset Information

print("\nDataset Information:")
df.info()


# Statistical Summary


print("\nStatistical Summary:")
print(df.describe())


# Unique Values in Categorical Columns


for col in categorical_features:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())


# Target Variable & Input Features


target_variable = "Survived"
input_features = df.drop(columns=[target_variable]).columns

print("\nTarget Variable:", target_variable)
print("Input Features:", list(input_features))

# Dataset Size


rows, cols = df.shape
print("\nNumber of Rows:", rows)
print("Number of Columns:", cols)


#  Data Quality Issues


print("\nMissing Values in Each Column:")
print(df.isnull().sum())

print("\nNumber of Duplicate Rows:")
print(df.duplicated().sum())

print("\nTarget Variable Distribution:")
print(df[target_variable].value_counts())
```
